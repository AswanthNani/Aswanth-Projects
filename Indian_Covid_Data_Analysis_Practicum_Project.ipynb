```python
#READING CSV FILE DATA
import pandas as pd
import numpy as np
import math
import matplotlib
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
covid = pd.read_csv(r"/covid_19_india.csv")
covid.head()
```

```python
#RENAMING COLUMNS AND GROUPING THE DATA
covid.rename(columns={'State/UnionTerritory':'State'},inplace=True)
covid=covid[["Date","State","Confirmed","Deaths","Cured"]]
grouped = covid.groupby(['State','Date'])
covid_confirmed = grouped.sum().reset_index().sort_values(['Date'],ascending=False)
covid_confirmed.head(36)
```

```python
#data columns are actually of ‘string’ datatype We need to convert them into ‘int’ datatype.
covid_confirmed['Confirmed'] = covid_confirmed['Confirmed'].map(int)
covid_confirmed['Deaths'] = covid_confirmed['Deaths'].map(int)
covid_confirmed['Cured'] = covid_confirmed['Cured'].map(int)
```

```python
df = (covid_confirmed[covid_confirmed['Date'].eq("2021-08-11")].sort_values(by="Confirmed",ascending=False).head(36))
df
```

```python
#Plotting A Bar Graph With 'State'vs 'Number Of Confirmed Cases'
fig, ax = plt.subplots(figsize=(25, 50))
ax.barh(df['State'], df['Confirmed'])
plt.xlabel('Number of Confirmed Cases')
plt.ylabel('State')
```

```python
#There are some drawbacks in the above graph ;for the states which have no/less covid cases, both asymptote and plots values had been overwritten
#Code For Overcoming The Drawback Faced above
fig, ax = plt.subplots(figsize=(25, 50))
def draw_barchart(date,case="Confirmed"):
    dff= (covid_confirmed[covid_confirmed['Date'].eq(date)].sort_values(by=case,ascending=False).head(36))[::-1]
    ax.clear()
    ax.barh(dff['State'], dff[case], color=["#980505","#CD1212","#D84E4E","#CB6262","#D39B5F","#F7EC10","#D0F710","#9CF710","#B4D67F","#969C8E"][::-1])
    dx = dff[case].max() / 200
    for i, (value, name) in enumerate(zip(dff[case], dff['State'])):
        ax.text(value-dx, i,     name,           size=14, weight=600, ha='right', va='bottom')
        ax.text(value-dx, i-.25, value, size=10, color='#17202A', ha='right', va='baseline')
        #ax.text(value+dx, i,     f'{value:,.0f}',  size=14, ha='left',  va='center')
    # polished styles
    ax.text(1, 0.4, date, transform=ax.transAxes, color='#777777', size=30, ha='right', weight=800)
    ax.text(0, 1.06, 'Number of Cases', transform=ax.transAxes, size=12, color='#777777')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0, 1.12, 'The number of '+str(case)+' Cases upto 11 August 2021',
            transform=ax.transAxes, size=24, weight=600, ha='left')
    ax.text(1, 0, 'by @Aswanth; credit @Aswanthkumar', transform=ax.transAxes, ha='right',
            color='#777777', bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
    plt.box(False)
    
draw_barchart('2021-08-11')
```

```python
from datetime import datetime
from matplotlib.dates import DateFormatter
```

```python
# Getting the state-wise daily data 
StateDailydf = pd.read_csv('https://api.covid19india.org/csv/latest/state_wise_daily.csv')
StateDailydf.fillna(0, inplace=True)
StateDailydf
```

```python
# Setting start and latest date
dfshape = StateDailydf.shape
BeginDate = datetime.strptime(StateDailydf['Date'][0],'%d-%b-%y').strftime("%d %B %Y")
EndDate = datetime.strptime(StateDailydf['Date'][dfshape[0]-1],'%d-%b-%y').strftime("%d %B %Y")
print(BeginDate)
print(EndDate)
```

```python
# Creating a dictionary for State codes as mentioned in above dataset
statesUT = {
    'TT': 'India',
    'AN':'Andaman and Nicobar Islands',
    'AP':'Andhra Pradesh',
    'AR':'Arunachal Pradesh',
    'AS':'Assam',
    'BR':'Bihar',
    'CH':'Chandigarh',
    'CT':'Chhattisgarh',
    'DN':'Dadra and Nagar Haveli',
    'DD':'Daman and Diu',
    'DL':'Delhi',
    'GA':'Goa',
    'GJ':'Gujarat',
    'HR':'Haryana',
    'HP':'Himachal Pradesh',
    'JK':'Jammu and Kashmir',
    'JH':'Jharkhand',
    'KA':'Karnataka',
    'KL':'Kerala',
    'LA':'Ladakh',
    'LD':'Lakshadweep',
    'MP':'Madhya Pradesh',
    'MH':'Maharashtra',
    'MN':'Manipur',
    'ML':'Meghalaya',
    'MZ':'Mizoram',
    'NL':'Nagaland',
    'OR':'Odisha',
    'PY':'Puducherry',
    'PB':'Punjab',
    'RJ':'Rajasthan',
    'SK':'Sikkim',
    'TN':'Tamil Nadu',
    'TG':'Telangana',
    'TR':'Tripura',
    'UP':'Uttar Pradesh',
    'UT':'Uttarakhand',
    'WB':'West Bengal'
}
```

```python
# Function to get 'confirmed' cases of a particular state from above dataset and making its 10- day moving average for our analysis
def fetchstate(state):
    
    Dateslist = pd.date_range(BeginDate, EndDate)

    stateindex = list(statesUT.keys())[list(statesUT.values()).index(state)]
    CCdata = []
    
    for i in range(dfshape[0]):
        if StateDailydf['Status'][i] == 'Confirmed':
            CCdata.append(StateDailydf[stateindex][i])

    df = pd.DataFrame(list(CCdata),index = Dateslist, columns = ['Confirmed'])
    df['MovingAverage'] = df.iloc[:,0].rolling(window=5).mean()
    df.fillna(0, inplace = True)
    print(state)
    print(df)
    return df
```

```python
# Function plots the data in columns of 3 of list of states passed as parameter
def PlotConfirmandMovingAverage(loc):

    noc = len(loc)
    nor = math.ceil(noc/3)

    fig, axes = plt.subplots(nrows = nor, ncols = 3, figsize = (25,18))
    fig.suptitle('Confirmed cases and 10-day Moving Average in India and its states as of ' + EndDate, fontsize = 20)
    plt.subplots_adjust(hspace = .4)
    
    k = 0
    for i in range(nor):
        for j in range(3):
            if k < noc:
                StateData = fetchstate(loc[k])
                
                axes[i,j].bar(StateData.index, StateData['Confirmed'], color = 'orange', width = 1, edgecolor = 'gray', linewidth = 0.05, label = 'Daily Confirmed Cases')
                axes[i,j].plot(StateData.index, StateData['MovingAverage'], color = 'blue', linestyle = '-', linewidth = 1,label = '10-Day Moving Average')
                axes[i,j].plot(StateData.index[-1], StateData['MovingAverage'][-1], color = 'blue', marker='o', markerfacecolor='blue', markersize=5)
                
                axes[i,j].set_title(loc[k] + ' (Total cases: ' + str(int(StateData['Confirmed'].sum())) + ')', 
                                    y = 1, fontsize = 15, fontweight='bold')
                
                axes[i,j].grid(axis = 'y', alpha = 0.5)
                
                myFmt = DateFormatter("%d-%b")
                axes[i,j].xaxis.set_major_formatter(myFmt)
                
                axes[i,j].spines['top'].set_visible(False)
                axes[i,j].spines['right'].set_visible(False)
                
                if k == noc-1:
                    handles, labels = axes[i,j].get_legend_handles_labels()
                
                k = k + 1
    
    fig.legend(handles, labels, bbox_to_anchor=(0.51, 0.95), loc='upper center', ncol = 3, fontsize = 15, frameon=False)
    plt.savefig('ConfirmandMovingAverage.png')
    
# --- Function Ends Here ---
```
